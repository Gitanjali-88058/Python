```python
inputs=df.drop('Salary_more_than_100k',axis='columns')
```

```python
target=df['Salary_more_than_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_s']=le_company.fit_transform(inputs['Company'])
inputs['job_s']=le_job.fit_transform(inputs['Job'])
inputs['degree_s']=le_degree.fit_transform(inputs['Degree'])
inputs.head()
```

```python
inputs_s=inputs.drop(['Company','Job','Degree'],axis='columns')
inputs_s
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_s,target)
```

```python
model.score(inputs_s,target)
```

```python
model.predict([[2,2,1]])
```

# Titanic.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
targets=df["Survived"]
targets=targets.head(20)
```

```python
inputs=df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns")
inputs.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
inputs['Sex_s']=le_Sex.fit_transform(inputs["Sex"])
```

```python
inputs.head(10)
```

```python
inputs=inputs.drop(['Sex'],axis='columns').head(20)
```

```python
inputs
```

```python
inputs=inputs.drop("Survived",axis="columns")
```

```python
inputs.Age=inputs.Age.fillna(5)
inputs
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,targets)
```

```python
model.score(inputs,targets)
```

```python
model.predict([[1,38,71,0]])
```

```python
model.predict([[1,38,71,1]])
```

```python
model.predict([[1,3,71,0]])
```
